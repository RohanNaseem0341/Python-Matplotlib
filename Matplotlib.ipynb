```python
import matplotlib.pyplot as plt

import pandas as pd

data=pd.read_csv("data.csv")
plt.plot(data['Duration'],data['Pulse'])
plt.show()
```

```python
plt.plot(data['Duration'],data['Pulse'],marker="*")
plt.show()
```

```python
plt.plot(data['Duration'],data['Pulse'],marker="*")

plt.xlabel('Duration')
plt.ylabel('Pulse')
```

```python
plt.plot(data['Duration'],data['Pulse'],marker="*")

plt.xlabel('Duration')
plt.ylabel('Pulse')
plt.title("Duration And Pulse Graph")
plt.show
```

```python
plt.plot(data['Duration'],data['Pulse'],marker="*",linestyle="solid",linewidth=3)#dashed, solid
plt.xlabel('Duration')
plt.xlabel('Pulse')
#for tilte
plt.title("Duration And Pulse Graph")
plt.show()
```
